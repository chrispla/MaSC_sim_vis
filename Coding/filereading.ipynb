```python
import glob
import numpy as np
import librosa
import os
```

```python
#File reading
all_dirs = []
arab = 0
swahili = 0
swahilidirs = []
for root, dirs, files in os.walk('./Dataset'): #change directory here
        for name in files:
            if '.wav' in name:
                filedir = os.path.join(root, name)
                all_dirs.append(filedir)
                if 'ArabMashriq' in filedir:
                    arab+=1
                elif 'EisenbergCollection' in filedir:
                    swahilidirs.append(filedir)
                    swahili+=1
file_no = len(all_dirs)
print('Arab = ' + str(arab))
print('Swahili = ' + str(swahili))
print(file_no)

swahililines = []
with open('music.csv') as f:
    for line in f:
        swahililines.append(line.split(','))
print(len(swahililines))
swahililines.pop(0)


swahilinames = []
for i in range(len(swahililines)):
    swahilinames.append(swahililines[i][3])
print(swahilinames)
print('HERE')
print(len(swahilidirs))
matching = 0
missing = []
for i in range(len(swahilinames)):
    if swahilinames[i] in swahilidirs:
        matching+=1
    else:
        missing.append(swahilinames[i])
print(len(missing))
print(missing)
```
